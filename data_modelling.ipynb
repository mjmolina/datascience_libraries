# Machine Learning

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import numpy as np
import xgboost as xgb
```

```python
# Read your data
file_name = pd.read_csv('IRIS.csv')
```

```python

# Load data into a Pandas DataFrame
iris_data = load_iris(as_frame=True)
file_name = iris_data.frame
file_name.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Define Features (X) and Target (y)
X = file_name.drop(columns=['species'])
y = file_name['species']

# Models need numeric targets (0, 1, 2) for multi-class classification
le = LabelEncoder()
y_encoded = le.fit_transform(y) 

# --- 2. Split Data (The same split is used for all models) ---
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, # Use encoded target
    test_size=0.3, 
    random_state=47872, 
    stratify=y_encoded # Ensures balanced classes
)

# --- 3. Model A: Scikit-learn K-Nearest Neighbors (KNN) ---
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train) 
knn_accuracy = accuracy_score(y_test, knn.predict(X_test))

# --- 4. Model B: Scikit-learn Logistic Regression ---
log_reg = LogisticRegression(max_iter=200, solver='liblinear') 
log_reg.fit(X_train, y_train)
log_reg_accuracy = accuracy_score(y_test, log_reg.predict(X_test))

# --- Results ---
print("--- Scikit-learn Model Comparison ---")
print(f"| Model | Accuracy |")
print(f"| :--- | :--- |")
print(f"| KNN (k=5) | {knn_accuracy:.4f} |")
print(f"| Logistic Regression | {log_reg_accuracy:.4f} |")
```


XGBoost (eXtreme Gradient Boosting) is an ensemble library optimized for speed and accuracy on tabular data. Its API is very similar to Scikit-learn's but often offers custom parameters for performance tuning.
